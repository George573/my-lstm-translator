```python
import pandas as pd
from BPETokenizer import BPETokenizer
```

```python
def load_data_csv(filename) -> tuple[list, list]:
    data = pd.read_csv(filename).dropna()
    return data['en'].astype(str).to_list(), data['fr'].astype(str).to_list()
```

```python
en_texts, fr_texts = load_data_csv('en-fr-1-mil.csv')
```

```python
en_tokenizer = BPETokenizer()
en_tokenizer.train(en_texts, n_merges=6_500)
```

```python
fr_tokenizer = BPETokenizer()
fr_tokenizer.train(fr_texts, n_merges=7_000)
```

```python
sample_text = "I am watching a beautiful sunset!"
encoded_tokens = en_tokenizer.encode(sample_text)
print(encoded_tokens)
```

```python
en_tokenizer.save('en_bpe_model.json')
fr_tokenizer.save('fr_bpe_model.json')
```
